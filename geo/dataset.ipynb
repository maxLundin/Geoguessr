```python
import os.path

import pickle
```

```python
def get_dataset_from_folder(path):
    files = []

    for filename in os.listdir(path):
        f = os.path.join(path, filename)
        if os.path.isdir(f):
            i = 0
            for image in os.listdir(f):
                files.append(f"{f}/{image}")
                i += 1

    filenames = list(map(lambda x: "".join(x.split("/")[1:]), files))

    print(filenames)
    classes = list(set(map(lambda x: x.split("/")[0], filenames)))
    numclasses = len(classes)
    classes = dict(zip(classes, range(numclasses)))

    return filenames, classes, numclasses
```

```python
filenames, classes, numclasses = get_dataset_from_folder('data/compressed_dataset')
```
