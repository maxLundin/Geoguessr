```python
import os.path
```

```python

directory = "data/compressed_dataset/"

filenames = []

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isdir(f):
        i = 0
        for image in os.listdir(f):
            filenames.append(f"{f}/{image}")
            i += 1
```

```python
fn1 = list(map(lambda x: "/".join(x.split("/")[2:]), filenames))
print(fn1)
```

```python
len(filenames)
```

```python
classes = {}

for file in filenames:
    file_class = file.split("/")[2]

    if file_class in classes:
        classes[file_class] += 1
    else:
        classes[file_class] = 1

print(len(classes))
```
